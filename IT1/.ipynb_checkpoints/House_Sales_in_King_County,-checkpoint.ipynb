### Transform data
Using sklearn base transformers
* ID number is useful
We can delete it to make our data more flexible and correct to next steps



```python
class DateReadableTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, date_column_name):
        self.date_column_name = date_column_name
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        new_date = X[self.date_column_name].apply(
                        lambda date_readable: date(int(date_readable[0:4]),
                                                   int(date_readable[4:6]),
                                                   int(date_readable[6:8])))
        X[self.date_column_name] = new_date
        return X
```

```python
date_transformer = DateReadableTransformer('date')
sales = date_transformer.fit_transform(sales)
```

```python
class FeatureDropTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, features_to_drop):
        self.features_to_drop = features_to_drop
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.drop(self.features_to_drop, axis=1)
```

```python
remove_id_transformer = FeatureDropTransformer('id')
sales = remove_id_transformer.fit_transform(sales)
```

### Sales over months
- It is not easy to get any pattern from the daily sales
- There's a good patern of sales average price over months though - Sales are the highest around April to June and then fall steadily until February, when they start to rise again
- I will create a transformer below that takes a dataset with a full date and changes it to months

```python
sales_date_month = deepcopy(sales)
sales_date_month['date'] = sales_date_month['date'].apply(lambda date_: date_.month)
sales_sorted_by_date = sales_date_month.sort_values(by='date', ascending=True)
sales_sorted_grouped_by_date_mean = sales_sorted_by_date[['price', 'date']].groupby(by='date').mean()
sales_sorted_grouped_by_date_median = sales_sorted_by_date[['price', 'date']].groupby(by='date').median()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))


axes[0].plot(sales_sorted_grouped_by_date_mean, label='mean', color='orange')
axes[0].legend()
axes[0].set_title('House Sales Price - Mean over time', fontsize=15)
axes[0].set_ylim(min(sales_sorted_grouped_by_date_mean['price']), 
                 max(sales_sorted_grouped_by_date_mean['price']))

axes[1].plot(sales_sorted_grouped_by_date_median, label='median')
axes[1].legend()
axes[1].set_title('House Sales Price - Median over time', fontsize=15)
axes[1].set_ylim(min(sales_sorted_grouped_by_date_median['price']), 
                 max(sales_sorted_grouped_by_date_median['price']))
fig.tight_layout()
```

```python
class DateToMonthTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, date_column_name='date'):
        self.date_column_name = date_column_name
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        # Checks if the date is not already transformed to numbers
        # if type(X['date'][0]) != np.int64:
        try:
            X['date'] = X['date'].apply(lambda date_: int(date_.month))
        except:
            pass
        return X

date_to_month_transformer = DateToMonthTransformer()
X_sales_train_date_months = date_to_month_transformer.fit_transform(X_sales_train_chosen)
```

## RMSE

```python
# Stochastic Gradient Descent Regressor 
#penalty = 'l2'
sgd_reg = SGDRegressor(penalty='l2', eta0=0.01, early_stopping=True,
                      random_state=95)

# Creating a for loop will let me save a lot of code
feature_names = X_sales_train_chosen_scaled.columns
rmse_list = []
counter = 1
fig, axes = plt.subplots(figsize=(15, 35))

for feature_name in feature_names:
    # *.reshape(-1, 1) is needed for single feature data, ax 2D array is expected
    current_feature = X_sales_train_chosen_scaled[f'{feature_name}'].values.reshape(-1, 1)
    sgd_reg.fit(current_feature, y_sales_train)
    # I am "reusing training data" here, test data is not pre-prepared
    sgd_reg_predictions = sgd_reg.predict(current_feature)
    current_model_rmse = round(np.sqrt(mean_squared_error(y_sales_train, sgd_reg_predictions)), 3)
    rmse_list.append(current_model_rmse)
    
    # Using X_sales_train_chosen - not the scaled version - to retain proper labels for xticks
    plt.subplot(7, 3, counter)
    plt.plot(X_sales_train_chosen[feature_name], y_sales_train,
             'bo', alpha=0.2)
    plt.title(label=feature_name+' - RMSE: '+str(current_model_rmse), fontsize=20)
    counter +=1    

plt.tight_layout()
sns.despine(left=True, bottom=True, top=True, right=True)    

feature_scores_names_sorted = sorted(zip(rmse_list, feature_names), reverse=False)
feature_scores_names_sorted
```

```python
# Mixed Selection - a small 'app' that informs me when the RMSE goes up
# with addition of some feature. Thanks to that I have insight what improves
# and what hurts the model. It also removes the feature from the features 
# to train on.

feature_names_current_training = []
current_model_rmse_mixed = []
current_model_features_mixed = []
previous_model_rmse = 0
current_model_rmse = 0

for rmse, feature_name in feature_scores_names_sorted:
    feature_names_current_training.append(feature_name)
    current_features = X_sales_train_chosen_scaled[feature_names_current_training]
    sgd_reg.fit(current_features, y_sales_train)
    sgd_reg_predictions = sgd_reg.predict(current_features)
    previous_model_rmse = current_model_rmse
    current_model_rmse = round(np.sqrt(mean_squared_error(y_sales_train, sgd_reg_predictions)), 3)
    print('CURRENTLY:    ', current_model_rmse, '   Feature added:', feature_name)
    if current_model_rmse >= previous_model_rmse and previous_model_rmse != 0:
        print('\tWarning! Feature name:', feature_name)
        print('\t\tCurrent:', round(current_model_rmse, 3), '>= Previous:', previous_model_rmse)
        print(f'\n\tRemoving the feature {feature_name} from the model')
        current_features = current_features.drop(feature_name, axis=1)
        feature_names_current_training.remove(feature_name)
        print('\t\t\t\t\t\t\tDONE')
        sgd_reg.fit(current_features, y_sales_train)
        sgd_reg_predictions = sgd_reg.predict(current_features)
        current_model_rmse = round(np.sqrt(mean_squared_error(y_sales_train, sgd_reg_predictions)), 3)
        print('\t\tRetraining...')
        print('\t\tWent back to RMSE: ', current_model_rmse)
        print('\t\t\t\t\t\t\tDONE\n\n')
    current_model_rmse_mixed.append(current_model_rmse)
    
sorted(current_model_rmse_mixed)
```

# Data Preparation Pipeline <a name="pipeline1"></a>
I am recreating here the data transformation for the following purposes:
- Readability
- The training data gets transformed - I need to transform the test data too (without looking into it) - Otherwise the models are going to be trained on training data that is different from the test data - thanks to using random_state I can do it one more time safely
- Model reusability

I also decide to keep the initial **yr_built** and **yr_renovated**, as **yr_built** is definitely a better feature, while **yr_ago_renovated** is ambiguous and the difference from **yr_renovated** is very small anyway.

```python
data_preprocessing_pipeline =  Pipeline([
    ('data_usable_format', DateReadableTransformer('date')),
    ('zipcode_dropper', FeatureDropTransformer(['zipcode', 'id'])),
    ('date_to_month', DateToMonthTransformer('date')),
    ('scaler', StandardScaler())
])
```

```python
sales = pd.read_csv("data/kc_house_data.csv")
sales = sales.drop(15870, axis=0) # Drop the 33 bedroom error datapoint
y_sales = sales['price']
X_sales = sales.drop(['price'], axis=1)

feature_names = list(X_sales.columns)
for feature in ['zipcode', 'id']:
    feature_names.remove(feature)
X_sales = data_preprocessing_pipeline.fit_transform(X_sales)
X_sales = pd.DataFrame(X_sales, columns=feature_names)
```

```python
X_sales_train, X_sales_test, y_sales_train, y_sales_test = train_test_split(
                                            X_sales, y_sales, test_size=0.2, random_state=95)
```

### Scoreboard & and a function that will add to it easily

```python
columns = ['Model Name', 'Description', 'train RMSE', 'test RMSE']
scoreboard = pd.DataFrame(columns=columns)
scoreboard
```

```python
def add_to_scoreboard(scoreboard_df, model_name, model_descr, train_rmse, test_rmse,
                      train_r2, test_r2, train_r2_adj='-', test_r2_adj='-'):
    data_to_append = pd.DataFrame.from_dict({'Model Name': [model_name], 
                                             'Description': [model_descr],
                                             'train RMSE': [train_rmse],
                                             'test RMSE': [test_rmse],})
    scoreboard = scoreboard_df.append(data_to_append)
    return scoreboard
```

### Create a function 'model_tests' <a name="model_tests"></a>
Testing the model of different regressions is quite similar.
- create the model
- reshape the size of an array to an acceptable form by sklearn if list of features contains only 1 feature
- allow for additional transformations like "polynomial features'
- calculate rmse
- add all the results to the scoreboard

```python
def model_tests(regressor, feature_list, scoreboard, model_name,
                model_description, additional_transformers=None):
    model = regressor
        
    if len(feature_list) == 1:
        X_features_train = X_sales_train[feature_list].values.reshape(-1, 1)
        X_features_test = X_sales_test[feature_list].values.reshape(-1, 1)
    else:
        X_features_train = X_sales_train[feature_list]
        X_features_test = X_sales_test[feature_list]

    if additional_transformers:
        for transformer in additional_transformers:
            X_features_train = transformer.fit_transform(X_features_train)
            X_features_test = transformer.fit_transform(X_features_test)
        
    n_samples_train = len(X_features_train)
    n_samples_test = len(X_features_test)
    n_features = len(np.asarray(X_features_train)[0])
    
    
    print('Number of samples train: ', n_samples_train,
          '\nNumber of samples test: ', n_samples_test,
          '\nNumber of features: ', n_features,
          '\nCurrent model: \n', model)
    
    model.fit(X_features_train, y_sales_train)

    # train statistics
    model_predictions_train = model.predict(X_features_train)
    model_train_rmse = round(mean_squared_error(y_sales_train, model_predictions_train)**0.5, 3)
    model_train_r2_score = r2_score(y_sales_train, model_predictions_train)
    model_train_r2_score_adj = r2_adjusted(model_train_r2_score,
                                           n_samples_train, n_features)

    # test statistics
    model_predictions_test = model.predict(X_features_test)
    model_test_rmse = round(mean_squared_error(y_sales_test, model_predictions_test)**0.5, 3)
    model_test_r2_score = r2_score(y_sales_test, model_predictions_test)
    model_test_r2_score_adj = r2_adjusted(model_test_r2_score,
                                           n_samples_test, n_features)
    if n_features == 1:
        model_train_r2_score_adj, model_test_r2_score_adj = '-', '-'

    scoreboard = add_to_scoreboard(scoreboard, model_name, model_description, model_train_rmse,
                                   model_test_rmse, model_train_r2_score, model_test_r2_score,
                                   model_train_r2_score_adj, model_test_r2_score_adj)
    return scoreboard, (regressor, model_name, model_description, model_train_rmse,
                        model_test_rmse, model_train_r2_score, model_test_r2_score,
                        model_train_r2_score_adj, model_test_r2_score_adj)
```

# Simple Linear Regression with 1 most important feature
### Final, correct approach <a name="simple_lin_reg"></a>
- As I can see from the scoreboard below, everything works as expected
- As said earlier, these results from the simplest model will serve as a benchmark for further models

```python
model = LinearRegression()
model_name = 'Linear Regression'
model_description = '1 feature, no regularization'
feature_list = ['sqft_living']
scoreboard, model_results = model_tests(model, feature_list, scoreboard,
                                          model_name, model_description)
scoreboard
```

# Polynomial Regression <a name="poly_reg"></a>

Adding transformations like powers to the regression works. It just requires adding the newly created features to the formula for cost function minimization.

I am going to use PolynomialFeatures from sklearn. It not only adds powers of the features, but also interaction terms. I have to be careful about so called combinatorial explosion, which happens when the degree of powers added is high and there are a lot of features. As can be seen below, going above 2nd degree does not improve the model so combinatorial explosion is not a concern.

<em>For some big datasets I could use SGDRegressor or create a MiniBatch Stochastic Regressor. I don't think there's a need for that as the dataset is relatively small.</em>

# Regularization<a name="regularization"></a>

Quickly on what regularization is: <em>"a form of regression, that constrains/ regularizes or shrinks the coefficient estimates towards zero. In other words, this technique discourages learning a more complex or flexible model, so as to avoid the risk of overfitting."</em>

Long story short, it is done by introducing a **regularization term** to the **cost function** during the training phase. A general very simplified formulas to explain it better could look like that:

1. **Not regularized:**    
$Mean\ Squared\ Error\ =\ some\ formula$
2. **Regularized:**     
$Mean\ Squared\ Error\ =\ some\ formula\ +\ \alpha*regularization\ term$


- The amount of regularization is controlled by hyperparameter alpha. When it is equal to zero, then the regularized regression becomes simple not regularized regression.

```python
# Visualisation based on "Hands On Machine Learning with Scikit Learn" 
# by Geron Aurelien

np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)

def plot_model(model_class, polynomial, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10,
                                                         include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw,
                 label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=4)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])

plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=95)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=95)

plt.show()
```
